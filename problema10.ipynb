```python
import matplotlib.pyplot as plt
from collections import Counter
from datetime import datetime
import matplotlib.dates as mdates
```

```python
def plot_appointments_by_day(appointment_dates):
    # Converter strings de data para objetos datetime para garantir formatação correta
    dates = [datetime.strptime(date, "%Y-%m-%d") for date in appointment_dates]
    
    # Contar a frequência de cada dia
    date_counts = Counter(dates)
    
    # Separar datas e contagens
    dates = list(date_counts.keys())
    counts = list(date_counts.values())
    
    # Ordenar as datas antes de plotar
    dates, counts = zip(*sorted(zip(dates, counts)))

    # Configurar estilo do gráfico
    plt.style.use('ggplot')

    # Criar o gráfico de barras
    fig, ax = plt.subplots(figsize=(15, 5))  # Aumento do tamanho da figura
    bars = ax.bar(dates, counts, color='skyblue', alpha=0.7, label='Número de Atendimentos')

    # Adicionar a linha no topo das barras
    ax.plot(dates, counts, color='darkred', marker='o', linestyle='-', linewidth=2, label='Tendência de Atendimentos')
    
    # Adicionar rótulos e título
    ax.set_xlabel('Data', fontsize=12, fontweight='bold')
    ax.set_ylabel('Número de Atendimentos', fontsize=12, fontweight='bold')
    ax.set_title('Quantidade de Atendimentos Médicos por Dia', fontsize=14, fontweight='bold')
    
    # Formatar o eixo x para mostrar datas mais claramente
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=2))  # Intervalo de dois dias
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Formato ano-mês-dia
    plt.xticks(rotation=90)  # Rotação de 90 graus

    # Melhorar layout e adicionar grade
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Apenas valores inteiros no eixo y

    # Adicionar legenda
    ax.legend()

    # Mostrar o gráfico
    plt.show()
```

```python
# Exemplo de entrada 1
appointment_dates = ["2023-09-01", "2023-09-01", "2023-09-02", "2023-09-03", "2023-09-03", "2023-09-03", "2023-09-05", "2023-09-07"]
plot_appointments_by_day(appointment_dates)
```

```python
# Exemplo de entrada com 15 atendimentos, 10 em 2024 e 5 em 2023
appointment_dates = [
    "2023-11-01", "2023-11-02", "2023-11-02", "2023-12-01", "2023-12-01",
    "2024-01-15", "2024-01-15", "2024-01-15", "2024-02-01", "2024-02-01",
    "2024-02-01", "2024-03-01", "2024-03-01", "2024-03-01", "2024-03-01"
]

plot_appointments_by_day(appointment_dates)
```
