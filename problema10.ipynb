```python
import matplotlib.pyplot as plt
from collections import Counter
from datetime import datetime, timedelta
import matplotlib.dates as mdates
import random
```

```python
def generate_random_dates(num_dates, year, month, day_range):
    # Gera uma lista de datas aleatórias dentro do mês especificado
    base_date = datetime(year, month, 1)
    return [base_date + timedelta(days=random.randint(0, day_range - 1)) for _ in range(num_dates)]

def plot_appointments_by_day(appointment_dates):
    # Conversor de strings de data para objetos datetime para garantir formatação correta
    dates = [datetime.strptime(date, "%Y-%m-%d") for date in appointment_dates]
    
    # Frequência de cada dia
    date_counts = Counter(dates)
    
    # Separar datas e contagens
    dates = list(date_counts.keys())
    counts = list(date_counts.values())
    
    # Ordenar as datas antes de plotar
    dates, counts = zip(*sorted(zip(dates, counts)))

    # Configurar estilo do gráfico
    plt.style.use('ggplot')

    # Criar o gráfico de barras
    fig, ax = plt.subplots(figsize=(12, 5))  # Ajuste do tamanho da figura
    bars = ax.bar(dates, counts, color='skyblue', alpha=0.7, label='Número de Atendimentos')
    
    # Adicionar a linha no topo das barras
    ax.plot(dates, counts, color='darkred', marker='o', linestyle='-', linewidth=2, label='Tendência de Atendimentos')
    
    ax.set_xlabel('Data', fontsize=12, fontweight='bold')
    ax.set_ylabel('Número de Atendimentos', fontsize=12, fontweight='bold')
    ax.set_title('Quantidade de Atendimentos Médicos', fontsize=14, fontweight='bold')

        # Formatar o eixo x para mostrar apenas as datas dos atendimentos
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    ax.set_xticks(dates)  # Definir ticks do eixo x diretamente com as datas de atendimentos
    plt.xticks(rotation=90)  # Rotação de 90 graus para melhor visualização
    

    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Apenas valores inteiros no eixo y
    ax.legend()
    plt.show()
```

```python
# Exemplo de entrada 1
appointment_dates = ["2023-09-01", "2023-09-01", "2023-09-02", "2023-09-03", "2023-09-03", "2023-09-03", "2023-09-05", "2023-09-07"]
plot_appointments_by_day(appointment_dates)
```

```python
# Exemplo de entrada com 15 atendimentos, 10 em 2024 e 5 em 2023
appointment_dates = [
    "2023-11-01", "2023-11-12", "2023-11-05", "2023-12-01", "2023-12-01",
    "2024-01-15", "2024-01-15", "2024-01-15", "2024-02-01", "2024-02-01",
    "2024-02-01", "2024-03-01", "2024-03-01", "2024-03-01", "2024-03-01"
]

plot_appointments_by_day(appointment_dates)
```

```python
#100 atendimentos no mês de junho de 2024
appointment_dates = generate_random_dates(100, 2024, 6, 30)
appointment_dates = [date.strftime("%Y-%m-%d") for date in appointment_dates]

plot_appointments_by_day(appointment_dates)
```
